**El Teorema  (Desigualdad de Chebyshev) mide la probabilidad de que cualquier variable aleatoria $X$ tome un valor dentro de $k$ desviaciones est√°ndar de la media es de al menos $\displaystyle \left(1-\frac{1}{k^2} \right)$**  

---
---    

    
$$
\begin{array}{rcl}
\displaystyle P\left( \mid X-\mu \mid > k \sigma \right) &\leq& \frac{1}{k^2} \\
\displaystyle P\left( \mid X-\mu \mid \leq k \sigma \right)&\geq& \left(1-\frac{1}{k^2}\right) \\ 
  &&  \\ 
\end{array}
$$
    
</font>


<span style="color:blue"> 
$$
\begin{array}{rcl}
 \displaystyle P(\mu-k\sigma \leq X \leq \mu+k\sigma) &\geq& \left(1-\frac{1}{k^2}\right)\\
 \displaystyle P(\mu-3\sigma \leq X \leq \mu+3\sigma) &\geq& \left(1-\frac{1}{3^2}\right)=0.88\\
 \displaystyle P(\mu-2\sigma \leq X \leq \mu+2\sigma) &\geq& \left(1-\frac{1}{2^2}\right)=0.75\\
\end{array}
$$

# Cuartil 
    
Los cuartiles son valores que dividen una muestra de datos en cuatro partes iguales. Utilizando cuartiles puede evaluar r√°pidamente la dispersi√≥n y la tendencia central de un conjunto de datos, que son los pasos iniciales importantes para comprender sus datos.

$$
\mbox{Cuartil	Descripci√≥n}\\
\begin{array}{|l|c|l|} \hline
\mbox{1er cuartil } &Q_1	&25\% \mbox{de los datos es menor que o igual a este valor.} \\ \hline
\mbox{2do cuartil } & Q_2	&50\% \mbox{de los datos es menor que o igual a este valor.} \\ \hline
\mbox{3er cuartil } &Q_3	&75\%   \mbox{de los datos es menor que o igual a este valor.} \\ \hline
\end{array}
$$
    
Observe adem√°s
    
*Primer cuartil (Q1) como la mediana de la primera mitad de valores*
    
*Segundo cuartil (Q2) como la propia mediana de la serie*
    
*Tercer cuartil (Q3) como la mediana de la segunda mitad de valores.*

**Rango intercuartil  (IQR)**	La distancia entre el primer 1er cuartil y el 3er cuartil (Q3-Q1); de esta manera, abarca el 50% central de los datos.
    
</font>

<font size="5">

---
- sf(x,$\mu$, $\sigma$)  :  1 - cdf(x)= $P(X \geq x)$

---    
- pmf(x,$\mu$, $\sigma$) : Funci√≥n de probabilidad f(x) (distribuciones discretas)

---
- pdf(x,$\mu$, $\sigma$) : Funci√≥n de densidad f(x) (distribuciones continuas)
    
---
- <span style="color:blue">  ppf(x,$\mu$, $\sigma$) : Funci√≥n inversa a cdf(x). </span> Nos permite obtener el valor correspondiente a una probabilidad.

## TEOREMA DEL L√çMITE CENTRAL 

</font>

---

<font size="4"> 
    
Si una variable aleatoria $Y$ es la suma de $n$ variables aleatorias independientes que satisfacen ciertas condiciones 
generales, entonces para $n$ suficientemente grande, $Y$ se encuentra aproximadamente **DISTRIBUIDA EN FORMA NORMAL**  
    


[TLC](https://support.minitab.com/es-mx/minitab/18/help-and-how-to/statistics/basic-statistics/supporting-topics/data-concepts/about-the-central-limit-theorem/)
    
</font>
</span> 
    
---
---

<span style="color:green">

<font size="4">


Si $X_1,X_2,\ldots,X_n$ es una secuencias de $n$ variables independientes con $\ \ E(X_i)=\mu_i$, $ \ \ \¬†V(X_i)=\sigma_i^2 \hspace{1cm} $ 
    
con $\ \ \ Y=X_1+X_2+\cdots+X_n \hspace{1cm}$ Entonces bajo ciertas condiciones generales se tiene:

$$Z_n=\frac{Y- \displaystyle \sum_{i=1}^n \mu_i}{ \sqrt{ \displaystyle \sum_{i=1}^n \sigma_i^2}}$$

Tiene una distribucion normal  $\ \ N(0,1) \ \$

</font>

<span style="color:blue">

### RESULTADO DIRECTO


<font size="4">
    
Si $X_1,X_2,\ldots,X_n$ es una secuencias de $n$ variables independientes e identicamente distribuidas *(i.i.d.)*, 
    
con $\ \ E(X_i)=\mu$, $ \ \ \¬†V(X_i)=\sigma^2 \ \ $

Entonces  $\hspace{1cm}Y=X_1+X_2+\cdots+X_n \hspace{1cm}$ 

\begin{array}{rcl}    
Z_n &=&\frac{Y- \displaystyle n \mu}{  \sqrt{ \displaystyle n \sigma^2}}\\
    & & \\
Z_n &=& \displaystyle \frac{\left(\frac{1}{n}\right)(Y- \displaystyle n \mu)}{\left(\frac{1}{n}\right)
    \left(\sqrt{ \displaystyle n \sigma^2 }\right)}\\
  & & \\
Z_n & = &\displaystyle \frac{\bar{Y}- \displaystyle \mu}{\displaystyle\sqrt{\frac{\sigma^2}{n}}}\\
    && \\
Z_n &=&
    \displaystyle \frac{\bar{y}- \displaystyle \mu}{\displaystyle \frac{\sigma}{\sqrt{n}}}
\end{array}    

Tiene una distribucion normal  $\ \ N(0,1) \ \$
